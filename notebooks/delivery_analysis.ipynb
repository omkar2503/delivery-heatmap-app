```python
import pandas as pd

# Load the uploaded CSV
df = pd.read_csv('cleaned_delivery_data_nagpur_nashik.csv')

# Group by Pincode and calculate metrics
pincode_summary = df.groupby('Pincode').agg(
    Total_Deliveries=('Package ID', 'count'),
    Successes=('Status', lambda x: (x == 'Delivered').sum()),
    Failures=('Status', lambda x: (x == 'Failed').sum()),
    Delays=('Status', lambda x: (x == 'Delayed').sum()),
    Avg_Delivery_Hours=('Delivery Time (hrs)', 'mean'),
    Most_Common_Reason=('Delay Reason', lambda x: x.mode()[0] if not x.mode().empty else 'NA'),
    Latitude=('Latitude', 'first'),
    Longitude=('Longitude', 'first')
)

# Calculate percentages
pincode_summary['Success Rate (%)'] = round((pincode_summary['Successes'] / pincode_summary['Total_Deliveries']) * 100, 2)
pincode_summary['Failure Rate (%)'] = round((pincode_summary['Failures'] / pincode_summary['Total_Deliveries']) * 100, 2)
pincode_summary['Delay Rate (%)'] = round((pincode_summary['Delays'] / pincode_summary['Total_Deliveries']) * 100, 2)

# Preview result
pincode_summary.head()
```

```python
# Group by City and Partner
partner_summary = df.groupby(['City', 'Partner']).agg(
    Total=('Package ID', 'count'),
    Success=('Status', lambda x: (x == 'Delivered').sum()),
    Avg_Delivery_Hours=('Delivery Time (hrs)', 'mean')
)

# Calculate Success Rate
partner_summary['Success Rate (%)'] = round((partner_summary['Success'] / partner_summary['Total']) * 100, 2)

# Preview the summary
partner_summary.reset_index().head()
```
